# Melakukan Imputasi

disini saya mengganti nilai yang hilang padakolom operating system version, dengan menyesuaikan os pada laptop yang ada, saya mengguanakan metode ini karena lebih efektif dibanding menggunakan mean karena jenis os yang berbeda, dan tidak saya hapus karena jumlah data yang hilang cukup banhyak

```python
df.loc[df["Operating System"] == "Windows", "Operating System Version"] = df.loc[df["Operating System"] == "Windows", "Operating System Version"].fillna("Windows 10")
df.loc[df["Operating System"] == "macOS", "Operating System Version"] = df.loc[df["Operating System"] == "macOS", "Operating System Version"].fillna("macOS Monterey")
df.loc[df["Operating System"] == "Linux", "Operating System Version"] = df.loc[df["Operating System"] == "Linux", "Operating System Version"].fillna("Ubuntu 22.04")
```

**3. Outliers**

```python
numerical_columns = ["Price", "Weight", "Screen Size"]

for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"Jumlah outliers di '{col}': {len(outliers)}")
    print(outliers[[col]])
```

# Construct Data

```python
def classify_device(row):
    if row["Category"] in ["Ultrabook", "Gaming", "Business", "Convertible"]:
        return "Laptop"
    elif row["Screen Size"] > 17.3 or row["Weight"] > 3.5:
        return "Desktop"

df["Device Type"] = df.apply(classify_device, axis=1)

print(df.head())
```

disini saya menambahkan kolom device type untuk membedakan laptop dan desktop
