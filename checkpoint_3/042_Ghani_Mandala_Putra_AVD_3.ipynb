# Melakukan Imputasi

disini saya mengganti nilai yang hilang padakolom operating system version, dengan menyesuaikan os pada laptop yang ada, saya mengguanakan metode ini karena lebih efektif dibanding menggunakan mean karena jenis os yang berbeda, dan tidak saya hapus karena jumlah data yang hilang cukup banhyak

```python
df.loc[df["Operating System"] == "Windows", "Operating System Version"] = df.loc[df["Operating System"] == "Windows", "Operating System Version"].fillna("Windows 10")
df.loc[df["Operating System"] == "macOS", "Operating System Version"] = df.loc[df["Operating System"] == "macOS", "Operating System Version"].fillna("macOS Monterey")
df.loc[df["Operating System"] == "Linux", "Operating System Version"] = df.loc[df["Operating System"] == "Linux", "Operating System Version"].fillna("Ubuntu 22.04")
```

**3. Outliers**

```python
df["Weight"].unique()
```

```python
df["Weight"] = df["Weight"].str.replace("kgs", "").str.replace("kg", "").astype(float)
```

```python
df["Weight"].unique()
```

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

outliers_price = detect_outliers_iqr(df, "Price")
outliers_weight = detect_outliers_iqr(df, "Weight")

len(outliers_price), outliers_price.head(), len(outliers_weight), outliers_weight.head()
```

# Construct Data

```python
df["Screen Size"] = df["Screen Size"].astype(str).str.replace('"', '', regex=False)

df["Screen Size"] = pd.to_numeric(df["Screen Size"], errors="coerce")
```

```python
def classify_device(row):
    if row["Category"] in ["Ultrabook", "Gaming", "Notebook", "Convertible"]:
        return "Laptop"
    elif row["Screen Size"] > 17.3 or row["Weight"] > 3.5:
        return "Desktop"
    else:
        return "Other"

# Menambahkan kolom baru "Device Type"
df["Device Type"] = df.apply(classify_device, axis=1)

# Menyimpan hasil ke file baru
df.to_csv("laptops_with_device_type.csv", index=False)

# Menampilkan beberapa baris pertama
print(df.head())
```

### Insight & Tindakan

Insight:
- Harga laptop lebih dipengaruhi oleh RAM & prosesor daripada storage.
- Laptop dengan RAM 8GB adalah yang paling populer di pasaran.
- Harga laptop cenderung naik seiring waktu, jadi pembelian lebih awal bisa lebih hemat.
- Retailer harus menyesuaikan strategi pemasaran dengan segmentasi pelanggan.
- Pelanggan sebaiknya memilih spesifikasi berdasarkan kebutuhan daripada hanya melihat harga.
- RAM lebih besar (16GB & 32GB) mulai meningkat dalam ketersediaannya, menunjukkan permintaan untuk laptop performa tinggi.
- Laptop dengan RAM 4GB atau lebih kecil semakin jarang ditemukan, menunjukkan bahwa spesifikasi rendah mulai ditinggalkan

Tindakan:
- Retailer sebaiknya fokus menjual laptop dengan RAM 8GB atau lebih karena paling banyak diminati.
- Produsen dapat meningkatkan produksi laptop dengan RAM 16GB atau lebih untuk memenuhi permintaan pasar yang berkembang ke arah performa tinggi.
- Retailer sebaiknya mengelompokkan produk berdasarkan segmentasi harga untuk memudahkan pelanggan memilih.
- Pelanggan bisa memilih laptop berdasarkan kebutuhan mereka daripada hanya berdasarkan harga.
- Pelanggan yang mencari laptop untuk kerja berat (editing, gaming) sebaiknya fokus pada RAM & prosesor daripada sekadar kapasitas penyimpanan.
